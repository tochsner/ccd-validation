```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = np.random.multivariate_normal(np.zeros(4), np.random.rand(4, 4), size=1000)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1])
```

```python
cov = np.cov(X.T)
```

```python
X.T.shape
```

```python
L = np.linalg.cholesky(np.linalg.inv(cov))
Z = (L.T @ X.T).T
```

```python
sns.scatterplot(x=Z[:, 0], y=Z[:, 1])
```

```python
L
```

```python
L_a = L[:2, :2]
X_a = X[:, :2]
```

```python
Z_a = (L_a.T @ X_a.T).T
```

```python
sns.scatterplot(x=X_a[:, 0], y=X_a[:, 1], )
```

```python
sns.scatterplot(x=Z_a[:, 0], y=Z_a[:, 1], )
```
