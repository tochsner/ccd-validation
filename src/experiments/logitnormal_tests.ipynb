```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from scipy import special


x = np.linspace(0, 1, 1000)
p = stats.norm(loc=2, scale=0.2).pdf(special.logit(x))/(x*(1-x))
```

```python
sns.scatterplot(x=x, y=p)
```

```python
np.argmax(p[1:-1])
```
