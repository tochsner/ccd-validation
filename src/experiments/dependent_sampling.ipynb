```python
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
mu = 2.0
beta = 0.1
sigma = 0.01

N = 1000000
```

```python
def sample_bp():
    return np.random.normal(1.0, 1.0, N)

def sample_b(bps):
    eps = np.random.normal(0, sigma, N)
    bs = mu - beta*bps + eps
    return bs
```

```python
bps = sample_bp()
bs = sample_b(bps)
```

```python
stats.pearsonr(bs, bps)
```

```python
sns.scatterplot(x=np.exp(bps), y=np.exp(bs))
```

```python
E_b = np.mean(bs)
E_bp = np.mean(bps)
E_bp_2 = np.mean(np.square(bps))
Var_b = np.var(bs)
Cov = np.cov(bs, bps)[0][1]

E_b, E_bp, E_bp_2, Var_b, Cov
```
