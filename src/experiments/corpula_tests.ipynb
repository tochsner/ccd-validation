```python
import numpy as np
import seaborn as sns
import scipy
```

```python
# generate a multivariate normal distribution with a given input correlation:
l = np.random.rand(2, 2)
mv = scipy.stats.multivariate_normal.rvs(mean = np.zeros(2), cov = l.T @ l, size = 5000)

# transform the sample to a uniform with the above correlation:
unif = scipy.stats.norm.cdf(mv)

# apply the marginal:
mv_fin = scipy.stats.beta.ppf(unif, a=5, b=1)

unif_inv = scipy.stats.beta.cdf(mv_fin, a=5, b=1)

mv_inv = scipy.stats.norm.ppf(unif_inv)
```
