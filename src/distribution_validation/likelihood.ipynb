```python
from pathlib import Path

import numpy as np
import pandas as pd


POSTERIOR_DIR = Path("data/distribution_data")
PLOTS_DIR = Path("plots/likelihood_plots")
```

```python
dict_likelihoods = {
    "dataset_name": [],
    "run": [],
    "model_name": [],
    "data_likelihood": [],
}

for posterior_file in list(POSTERIOR_DIR.glob("*.log")):
    file_name_wo_ext = posterior_file.name.removesuffix(".log")
    dataset_name, run, _, model_name = file_name_wo_ext.split("_")

    likelihoods = pd.read_csv(posterior_file)

    dict_likelihoods["dataset_name"].append(dataset_name)
    dict_likelihoods["run"].append(run)
    dict_likelihoods["model_name"].append(model_name)
    dict_likelihoods["data_likelihood"].append(
        np.sum(np.log(likelihoods.posterior))
    )

df_likelihoods = pd.DataFrame(dict_likelihoods)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for dataset_name, df_likelihoods_per_dataset in df_likelihoods.groupby("dataset_name"):
    df_likelihoods_per_dataset = df_likelihoods_per_dataset.sort_values("data_likelihood")
    
    sns.boxplot(
        df_likelihoods_per_dataset,
        x="model_name",
        y="data_likelihood",
    )

    plt.title(f"Data Likelihood ({dataset_name})")
    plt.xlabel("Model")
    plt.ylabel("Data Likelihood")
    
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()
    
    plt.savefig(PLOTS_DIR / f"{dataset_name}_data-likelihood.png", dpi=300)
    plt.close()

    df_wins = df_likelihoods_per_dataset.sort_values("data_likelihood", ascending=False).drop_duplicates(["run"])
    sns.countplot(
        df_wins,
        x="model_name",
    )
    
    plt.title(f"Data Likelihood Wins ({dataset_name})")
    plt.xlabel("Model")
    plt.ylabel("Number of Runs")

    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()
    
    plt.savefig(PLOTS_DIR / f"{dataset_name}_wins.png", dpi=300)
    plt.close()
```
