```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid")
```

```python
df = pd.DataFrame(
    {
        "method": ["Prior+1", "Prior+1", "Prior+1", "Prior+1", "CCD0", "CCD0", "CCD0", "CCD0"],
        "sample_size": [100, 100, 200, 200, 100, 100, 200, 200],
         "rf": [3, 5, 2, 4, 4, 7, 5, 3],
    }
)
df
```

```python
sns.lineplot(
    data=df,
    x="sample_size",
    hue="method",
    y="rf",
)

plt.xlabel("Sample size")
plt.ylabel("RF")
plt.title("RF Distance to True Tree")
```

```python
sns.catplot(
    data=df,
    x="sample_size",
    hue="method",
    y="rf",
    kind="point",
    errorbar="se",
    capsize=0.1
)

plt.xlabel("Sample size")
plt.ylabel("RF")
plt.title("RF Distance to True Tree")
```
